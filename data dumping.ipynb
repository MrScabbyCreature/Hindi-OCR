```python
import os
import numpy as np
import pickle
import cv2
import pandas as pd
import matplotlib.pyplot as plt

def display_image(image, title=''):
    cv2.imshow(title, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    

def display_all_in_a_list(image_list, titles=None, scale_factor=1):
    '''
    image_list: a single image, or a list of images
    -----------------------------------
    uses matplotlib to display all images in 'image_list' in the same cell
    '''
    plt.rcParams['figure.figsize'] = [10*scale_factor, len(image_list)*5*scale_factor]
    _, axes = plt.subplots(len(image_list), 1)
    if len(image_list) == 1:
        axes.imshow(image_list[0])
        if titles is not None:
            axes.set_title(titles[0])
    else:
        for index in range(len(image_list)):
            axes[index].imshow(image_list[index])
            if titles is not None:
                axes[index].set_title(titles[index])
    plt.show()
```

```python
folder = 'base_characters'
len(os.listdir(folder))
```

```python
data = {}
for character_name in os.listdir(folder):
    print(character_name, len(os.listdir(os.path.join(folder, character_name))))
#     data[character_name] = [cv2.threshold(cv2.imread(folder+'/'+character_name+'/'+file, 0),0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1] \
#                             for file in os.listdir(folder+'/'+character_name)]
    data[character_name] = [cv2.imread(folder+'/'+character_name+'/'+file, 0)/255 \
                            for file in os.listdir(folder+'/'+character_name)]
```

```python
for ch in data:
    for i in range(len(data[ch])):
        data[ch][i] = data[ch][i][:, (data[ch][i]/255).sum(axis=0)!=28]
        data[ch][i] = data[ch][i][(data[ch][i]/255).sum(axis=1)!=28, :]
    for i in range(5):
#         if data[ch][i].shape[1] < 28 or data[ch][i].shape[0] < 28:
        print(ch, data[ch][i].shape)
        display_all_in_a_list([data[ch][i]])
```
