```python
import os
import pandas as pd
import numpy as np
from IPython.display import display as display_dataframe
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
folder = "results/part classifiers/"
os.chdir(folder)
```

```python
def separate_characters_and_accuracy(folder):
    files = os.listdir(folder)
    if better_folder in files:
        files.remove(better_folder)
    if worse_folder in files:
        files.remove(worse_folder)
    files = list(map(lambda x: x[:-5], files))
    files = list(map(lambda x: (x.split('_')[:-1], x.split('_')[-1]), files))
    df = pd.DataFrame(files, columns=['char', 'acc'])
    df['acc'] = df['acc'].astype(np.float64)
    return df

better_folder = "better_ones"
worse_folder = "worse_ones"

# display(separate_characters_and_accuracy('2'))

plot_data = []
results = pd.DataFrame(columns = ['num', 'char', 'acc'])

for i in range(2, 37):
    all_frame = separate_characters_and_accuracy(str(i))
    better_frame = separate_characters_and_accuracy(os.path.join(str(i), better_folder))
    worse_frame = separate_characters_and_accuracy(os.path.join(str(i), worse_folder))
    
#     a = worse_frame[worse_frame['acc'] == worse_frame['acc'].min()]
    a = better_frame[better_frame['acc'] == better_frame['acc'].min()]
    a['num'] = i
    
    results = results.append(a, ignore_index=True)
#     print(i)
#     print(np.average(all_frame['acc']))
#     display(better_frame[better_frame['acc'] == better_frame['acc'].max()])
#     display(worse_frame[worse_frame['acc'] == worse_frame['acc'].min()])
#     print('\n\n')
    plot_data.append((i, np.average(all_frame['acc']), np.max(all_frame['acc']), np.min(all_frame['acc'])))
```

```python
pd.set_option('max_colwidth',400)
results = results[['num', 'char', 'acc']]
results
```

```python
plot_data = pd.DataFrame(plot_data, columns=['num', 'avg', 'max', 'min'])
plot_data = plot_data.set_index("num")
display_dataframe(plot_data)
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(data=plot_data)
plt.xlabel("Number of Classes")
plt.ylabel("Accuracy")
plt.savefig("main_graph.png")
```
