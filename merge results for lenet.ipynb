```python
a = []

for i in range(2, 37):
    a.append((i, np.average(results[results['Label_count'] == i]['Test_acc']), np.average(results[results['Label_count'] == i]['Overall_acc'])))

# pd.DataFrame(list(zip(list(range(10)), list(range(10, 20)))))
a = pd.DataFrame(a, columns = ['Num classes', 'Test', 'Overall'])
a = a.set_index('Num classes')
a['Train'] = (a['Overall'] - 0.2*a['Test'])/0.8
a[['Test', 'Train']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lineplot(data=a[['Test', 'Train']])
plt.xlabel("NoC")
plt.ylabel("Accuracy")
plt.savefig("accuracy_graph.png")
```
